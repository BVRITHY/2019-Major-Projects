```python
import pandas as pd
import numpy as np

import cv2
import os
```

```python
data = pd.read_csv("corn/train.csv")
data
```

```python
broken = []
pure = []
silkcut = []
discolored = []

for image,label in zip(data["image"],data["label"]):
    img = image.split('/')
    if label == "broken":
        broken.append(img[1])
    elif label == "pure":
        pure.append(img[1])
    elif label == "discolored":
        discolored.append(img[1])
    else:
        silkcut.append(img[1])

print(pure)
print(broken,silkcut,discolored)
```

```python
img = cv2.imread("corn/train/"+pure[0])
path = "dataset/pure/image.png"
cv2.imwrite(path,img)
```

```python
main_folder = "dataset"
try:
    os.mkdir(main_folder+"/pure")
    os.mkdir(main_folder+"/discolored")
    os.mkdir(main_folder+"/silkcut")
    os.mkdir(main_folder+"/broken")
except:
    pass
```

```python
count = 0
for i in pure:
    img = cv2.imread("corn/train/"+i)
    
    path = f"dataset/pure/{count}.png"
    count += 1
    cv2.imwrite(path,img)

count = 0
for i in discolored:
    img = cv2.imread("corn/train/"+i)
    
    path = f"dataset/discolored/{count}.png"
    count += 1
    cv2.imwrite(path,img)

count = 0
for i in silkcut:
    img = cv2.imread("corn/train/"+i)
    
    path = f"dataset/silkcut/{count}.png"
    count += 1
    cv2.imwrite(path,img)

count = 0
for i in broken:
    img = cv2.imread("corn/train/"+i)
    
    path = f"dataset/broken/{count}.png"
    count += 1
    cv2.imwrite(path,img)

categories = os.listdir(main_folder)
try:
    categories.remove(".DS_Store")
except:
    pass

print(categories, len(categories))
```
